```python
def generate_lines(filePath, delimiters=[]):
    with open(filePath) as f:
        for line in f:
          line = line.strip() #removes newline character from end
          for d in delimiters:
            line =line.replace(d, ' ')
          yield line
```

```python
def LanternFish(data, days):
  for i in range(days):
    newborn = [8]*(data == 0).sum()
    data -= 1
    data[data == -1] = 6
    data = np.append(data, newborn)
  return len(data)
```

```python
def LanternFish2(data, days):
  fishlist = np.zeros(9)
  for i in data:
    fishlist[i] +=1
  for i in range(days):
    fishlist[7]+=fishlist[0]
    fishlist = fishlist[[1,2,3,4,5,6,7,8,0]]
  return fishlist
```

```python
#Part 1
giant_list0 = []
giant_list1 = []
for i in content:
  smol_list = i.split("|")
  giant_list0.append(smol_list[0].split())
  giant_list1.append(smol_list[1].split())

def count1478(giant_list1):
  count = 0
  for i in giant_list1:
    for j in i:
      if len(j) == 2 or len(j) == 3 or len(j) == 4 or len(j) == 7:
        count +=1
  return count

print(count1478(giant_list1))
```

```python
#Part 2 
def detect_number(list0): #this list is just a list of strings
  lol = []
  number = {}
  for i in list0:
    lol.append(list(i))
    if len(i) == 2:
      number[1] = list(i) 
      lol.remove(list(i))
    elif len(i) == 3:
      number[7] = list(i)
      lol.remove(list(i))
    elif len(i) == 4:
      number[4] = list(i)
      lol.remove(list(i))
    elif len(i) == 7:
      number[8] = list(i)
      lol.remove(list(i))
  
  for i in lol:
    if len(i) == 5:
      check =  all(e in i for e in number[1])
      if check == True:
        number[3] = i
        lol.remove(i)

  for i in lol:
    if set(i) == (set(number[3]) | set(number[4])):
      number[9] = i
      lol.remove(i)

  for i in lol:
    if set(number[9]) == (set(number[1]) | set(i)):
      number[5] = i
      lol.remove(i)

  for i in lol:
    if len(i) == 5:
      number[2] = i
      lol.remove(i)
  
  for i in lol:
    if set(number[8]) == (set(number[1]) | set(i)):
      number[6] = i
      lol.remove(i)
  
  number[0] = lol[0]
  return number
```

```python
def whatnumber(string, dic):
  l = list(string)
  num = -1
  for i in dic:
    if set(l) == set(dic[i]):
      num = i
  return num

def addingup(list0,list1):
  sum = 0
  for i in range(len(list0)):
    dic = detect_number(list0[i])
    for j in range(len(list1[i])):
      num = whatnumber(list1[i][j], dic)
      sum += 10**(3-j)*num
  return sum

print(addingup(giant_list0,giant_list1))
```

##Day 9: Smoke Basin

```python
filePath = '/content/drive/MyDrive/AdventOfCode/input9.txt'

filename = open(filePath, "r")
G = []
gr = filename.readline().strip()
while gr:
    gr = list(map(int,str(gr)))
    G.append(gr)
    gr = filename.readline().strip()

G = np.array(G)
G = np.pad(G, pad_width=1, mode='constant', constant_values=9)

def findlow(G):
  coord = []
  sum = 0
  for i in range(1, len(G)-1):
    for j in range(1, len(G[i])-1):
      if G[i][j] < G[i-1][j] and G[i][j] < G[i][j-1] and G[i][j] < G[i+1][j] and G[i][j] < G[i][j+1]:
        coord.append([i,j])
        sum = sum + G[i][j] + 1

  return sum,coord

print(findlow(G))
```

```python
def sizeofpit(coordinate,G):
  Q = [coordinate]
  for i in Q:
    x,y = i[0],i[1]
    if G[x-1][y]!=9 and [x-1,y] not in Q:
      Q.append([x-1,y])
    if G[x][y-1]!=9 and [x,y-1] not in Q:
      Q.append([x,y-1])
    if G[x+1][y]!=9 and [x+1,y] not in Q:
      Q.append([x+1,y])
    if G[x][y+1]!=9 and [x,y+1] not in Q:
      Q.append([x,y+1])
  #print(Q)
  return(len(Q))
```

```python
def threelargest(G):
  risk, coords = findlow(G)
  sizes = []
  for i in coords:
    sizes.append(sizeofpit(i,G))
  sizes.sort(reverse=True)

  prod = sizes[0]*sizes[1]*sizes[2]
  return prod

print(threelargest(G))
```

```python
filePath = '/content/drive/MyDrive/AdventOfCode/input10.txt'

file1 = open(filePath, "r")
content = file1.read().splitlines()
file1.close()

print(content)
```

```python
test = "{([(<{}[<>[]}>{[]{[(<()>"
test = list(test)
print(test)

def corrupt(lista):
  Q = []
  left = ['{', '(', '[','<']
  right = ['}', ')', ']', '>']
  for i in lista:
    if i in left:
      Q.append(i)
    elif i in right:
      match_left = left.index(Q[-1]) # index of the character in left
      match_right = right.index(i)
      if match_left == match_right:
        Q.pop(-1)
      else:
        return i
  

print(corrupt(test))
```
